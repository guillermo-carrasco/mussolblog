## Collecting metrics for disk usage
[SAR](http://www.ibm.com/developerworks/aix/library/au-unix-perfmonsar.html) is a tool to get usage metrics from all the resources in your system.

Here I present a method to parse the disk usage file in order to plot it for a better visualization.

##Parsing the file

```python
with open('metrics/disk.txt', 'r') as f:
    metrics = f.readlines()
```

All the collected txt files start with this line:

```python
print metrics[0]
```

And the first line of data contains all the labels:

```python
print metrics[2]
```

DEV indicate the device-partition from where the metrics are taken, so we end up with a line per device:

```python
for l in metrics[3:21]:
    print l
```

We can save the dava in different dictionaries. Each dictionary will contain as a key, the dev-partition string, and as values, tuples containing (time, metrics). We can later plot this dictionaries easily:

```python
labels = metrics[2].split()[1:]
y, m, d = metrics[0].split()[3].split('-')
print labels, y, m, d
```

```python
import string
import datetime

devs = {}
for l in metrics[3:]:
    metrics_line = l.split()
    dev = metrics_line[1]
    t = metrics_line[0]
    data = metrics_line[2:]
    #We have to convert the timestamps to python datetime objects, and comma
    #separated floats to point separated floats
    t = datetime.datetime(int(y), int(m), int(d), int(t.split(':')[0]), int(t.split(':')[1]), int(t.split(':')[2]))
    for i in range(len(data)):
        data[i] = string.atof(data[i].replace(',', '.'))
    #And then we save it
    if devs.has_key(dev):
        devs[dev].append((t, data))
    else:
        devs[dev] = [(t, data)]
```

```python
#These are our devices
devs.keys()
```
